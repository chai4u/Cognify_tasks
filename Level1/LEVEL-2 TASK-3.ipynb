```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as pyplot
%matplotlib inline
import seaborn as sns
df=pd.read_csv("/content/drive/MyDrive/Dataset .csv")
```

```python
df['Restaurant Name Length'] = df['Restaurant Name'].apply(lambda x:len(x))
df['Address Length'] = df['Address'].apply(lambda x:len(x))
print(df)
```

```python
df['Restaurant Name List'] = df['Restaurant Name'].apply(lambda x:list(x))
df['Address List'] = df['Address'].apply(lambda x:list(x))
print(df)
```

```python
df['Restaurant Name List'] = df['Restaurant Name'].apply(lambda x:enumerate(x))
df['Address List'] = df['Address'].apply(lambda x:enumerate(x))
print(df)
```

```python
df['Restaurant Name List'] = df['Restaurant Name'].apply(lambda x:frozenset(x))
df['Address List'] = df['Address'].apply(lambda x:frozenset(x))
print(df)
```

```python
df['Has Table booking']=df['Has Table booking'].map({'Yes':1,'No':0})
print(df)
```
