```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Dataset .csv")
totalrestaurants_by_price_range=df["Price range"].value_counts()
onlinedelivery_counts=df['Has Online delivery'].value_counts()
percentage_onlinedelivery_by_pricerange = {}
for pricerange in totalrestaurants_by_price_range.index:
      subset_df = df[df["Price range"] == pricerange]
      onlinedelivery_counts = subset_df['Has Online delivery'].value_counts()
      percentage_onlinedelivery = {onlinedelivery_counts.get('Yes',0)/totalrestaurants_by_price_range[pricerange]*100}
      percentage_onlinedelivery_by_pricerange[pricerange] = percentage_onlinedelivery

print("percentage of restaurants offering online delivery by price range:")
print(percentage_onlinedelivery_by_pricerange)
```
