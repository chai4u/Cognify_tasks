```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as pyplot
%matplotlib inline
import seaborn as sns
df=pd.read_csv("/content/drive/MyDrive/Dataset .csv")
```

```python
#mcpr(most common price range)
pricerange_counts = df['Price range'].value_counts()
mcpr = pricerange_counts.idxmax()
print("most common price range among all the restaurants:",mcpr )
```

```python
average_rating_by_pricerange=df.groupby("Price range")["Aggregate rating"].mean()
print("Average rating for each price range:")
print(average_rating_by_pricerange)
```

```python
average_rating_by_pricerange=df.groupby("Price range")['Aggregate rating'].mean()
highest_average_rating_by_pricerange=average_rating_by_pricerange.idxmax()
highest_rating_rows=df[df['Price range'] == highest_average_rating_by_pricerange]
highest_avereage_rating_color=highest_rating_rows['Rating color'].iloc[0]
print('color that represents the highest average range:',highest_avereage_rating_color)
```
